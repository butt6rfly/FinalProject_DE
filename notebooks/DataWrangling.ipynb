## 2. Data Cleaning

### 2.1 Standardize the column names

```python
dirty.columns = [col.strip().lower().replace(' ', '_').replace('-','_') for col in dirty.columns]
```

```python
dirty.columns
```

### 2.2 Categorical columns

```python
categorical = dirty.select_dtypes(include='str').columns
categorical
```

```python
dirty.columns
```

#### Categorical columns to numeric

```python
categorical_to_numeric = ['age','detailed_industry_recode','wage_per_hour','capital_gains', 'weeks_worked_in_year']
```

```python
dirty[categorical] = dirty[categorical].fillna('NaN') #para nao dar erro no split
```

```python
for col in categorical_to_numeric:
    dirty[col] = [n.split(sep=' ')[0] for n in dirty[col]]
```

```python
for col in list(categorical_to_numeric):
    print(col, '\n', dirty[col].unique(), end='\n\n')
```

```python
replacements_categorical_to_numeric = {
    'unknown' : np.nan,
    'NaN' : np.nan,
    '' : np.nan,
    '?' : np.nan
}
```

```python
for col in categorical_to_numeric:
    dirty[col] = pd.to_numeric(dirty[col].replace(replacements_categorical_to_numeric), errors='coerce')

dirty[categorical_to_numeric].describe()
```

#### Replacing values

```python
replacements_categorical = {
    'NaN' : np.nan,
    ' ?' : np.nan,
    'unknown' : np.nan,
    'na' : np.nan
}
```

```python
for col in list(dirty.select_dtypes('str').columns):
    dirty[col] = dirty[col].replace(replacements_categorical).str.strip().str.lower().str.replace(r'[\s-]+', '_', regex=True)
    print(col, '\n', dirty[col].unique(), end='\n\n')
```

```python
replacements_dict = {
#class_of_worker 
    'self-employed-not-inc' : 'self_employed_not_incorporated',
    'self_employed_not_inc' : 'self_employed_not_incorporated',
    'not_in_universe' : np.nan,
#education
    'bachelors' : 'bachelors_degree(ba_ab_bs)',
    "bachelor's" : 'bachelors_degree(ba_ab_bs)',
    'hs_grad' : 'high_school_graduate',
#marital_stat
    'married-a_f_spouse_present' : 'married_a_f_spouse_present',
#sex 
    'm' : 'male',
    'f' : 'female',
#detailed_household_and_family_stat 
    'other_rel_<18_never_married_rp_of_subfamily' : 'other_rel_<18_never_marr_rp_of_subfamily',
#detailed_household_summary_in_household 
    'child_under_18_ever_married' : 'child_under_18_never_married',
#live_in_this_house_1_year_ago 
    'not_in_universe_under_1_year_old': 'no',
#country_of_birth_father 
    'trinadad&tobago' : 'trinidad_and_tobago',
#income 
    '50000+.' : '>50000',
    '_50000.' : '<=50000',
#source_split 
    'training' : 'train',
    'testing' : 'test'
    }
```

```python
for col in list(dirty.select_dtypes('str').columns):
    dirty[col] = dirty[col].replace(replacements_dict)
    print(col, '\n', dirty[col].unique(), end='\n\n')
```

```python
(dirty.select_dtypes('str').isna().sum()/len(dirty)*100).sort_values(ascending=False)
```

```python
dirty.select_dtypes('str').fillna('no_information')
```

#### Missing Values

```python
(dirty.select_dtypes('str').isna().sum()/len(dirty)*100).sort_values(ascending=False)
```

```python
for col in dirty.select_dtypes('str').columns:
    dirty[col] = dirty[col].fillna('no_information')
```

```python
(dirty.select_dtypes('str').isna().sum()/len(dirty)*100).sort_values(ascending=False)
```

### 2.3 Numerical columns

```python
dirty.select_dtypes('number').describe()
```

```python
(dirty.select_dtypes('number').isna().sum()/len(dirty)*100).sort_values(ascending=False)
```

```python
for col in dirty.select_dtypes('number').columns:
    dirty[col] = dirty[col].fillna(dirty[col].median())
```

```python
(dirty.select_dtypes('number').isna().sum()/len(dirty)*100).sort_values(ascending=False)
```
