### Missing values

```python
((df.isna().sum())/len(df)*100).sort_values(ascending=False)
```

```python
#Colunas com mais de 50% de valores nulos
df = df.drop(columns=['fill_inc_questionnaire_for_veteran_admin','reason_for_unemployment','enroll_in_edu_inst_last_wk','state_of_previous_residence','region_of_previous_residence','migration_prev_res_in_sunbelt','member_of_a_labor_union','family_members_under_18','class_of_worker','migration_code_change_in_reg','migration_code_move_within_reg','migration_code_change_in_msa','major_occupation_code'])
```

```python
((df.select_dtypes('number').isna().sum())/len(df)*100).sort_values(ascending=False)
```

```python
for col in df.select_dtypes('number').columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
((df.isna().sum())/len(df)*100).sort_values(ascending=False)
```

```python
for col in df.select_dtypes('str').columns:
    df[col] = df[col].fillna(df[col].mode().iloc[0])
```

```python
((df.isna().sum())/len(df)*100).sort_values(ascending=False)
```
